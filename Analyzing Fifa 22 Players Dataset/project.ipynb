1. Importing neccessary libraries viewing and preprocessing data

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
pd.set_option("display.width", 1000)
pd.set_option("display.max_rows", 1000)
pd.set_option("display.max_columns", 1000)
```

2. importing dataset

```python
df = pd.read_csv("players_21.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
col = ['player_face_url','club_logo_url','club_flag_url','nation_logo_url','nation_flag_url']
df1 = df.drop(columns=col, axis=1)
df1.shape
```

#### Date Conversion into year only

```python
df1['club_joined'] = pd.to_datetime(df1['club_joined'])
df1['joined_year'] = df1['club_joined'].dt.year
df1.head()
```

#### Exploratory Analysis

1. Plotting distribution of overall rating players

```python
sns.displot(df1['overall'], bins=20, color='r')
plt.xticks(rotation=45)
plt.title("Distribution of Overall ratings of all Players")
tick_values = list(range(50, 99, 5))  # Adjust the range and step as needed
plt.xticks(tick_values)
plt.savefig("Distribution.png")
plt.show()
```

2. AGE DISTRIBUTION

```python
sns.displot(df1['age'], bins=10, color='r')
plt.xticks(rotation=45)
plt.title("Distribution of Overall ratings of all Players") 
plt.savefig("Distribution.png")
plt.show()
```

```python
age_counts = df1['age'].value_counts().reset_index()
age_counts.columns = ['Age', 'Count']

# Display the summary table
print(age_counts)
```

3. pair plot

```python
sns.set_theme()
cols = ['overall', 'value_eur', 'wage_eur', 'international_reputation', 'height_cm', 'weight_kg', 'release_clause_eur'] 
sns.pairplot(df1[cols], height = 2.5)
plt.savefig("myplot.png")
plt.show()
```
